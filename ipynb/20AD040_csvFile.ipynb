```python
import matplotlib.pyplot as plt
import pandas as pd
csv=pd.read_csv('ADL_202_ex.csv')
mean_mark=csv['Mark'].mean()

#sorting data
#males
maleData=csv.loc[csv['Gender'] == 'm']
maleMean = maleData['Mark'].mean()
maleLow= maleData['Mark'].min() 
maleHigh= maleData['Mark'].max()
maleToppers=maleData.loc[(maleData['Mark']==maleHigh)]
maleLowers=maleData.loc[(maleData['Mark']==maleLow)]
#females
femaleData=csv.loc[csv['Gender'] == 'f']
femaleMean = femaleData['Mark'].mean()
femaleLow = femaleData['Mark'].min() 
femaleHigh= femaleData['Mark'].max() 
femaleToppers=femaleData.loc[(femaleData['Mark']==femaleHigh)]
femaleLowers=femaleData.loc[(femaleData['Mark']==femaleLow)]

#graph plotting
plt.title("Mark Distribution")
plt.xlabel("Sl No")
plt.ylabel("Marks")

#males and females mark scatter graph
plt.scatter(maleData['Sl No'],maleData['Mark'],label="Marks of Males")
plt.scatter(femaleData['Sl No'],femaleData['Mark'],label="Marks of Females")
plt.axhline(mean_mark,label="Overall Mean",color='g',linestyle='dotted')

#high and low marks
plt.scatter(maleToppers['Sl No'],maleToppers['Mark'],label="Male Toppers",color='b')
plt.scatter(femaleToppers['Sl No'],femaleToppers['Mark'],label="Female Toppers",color='r')
plt.scatter(maleLowers['Sl No'],maleLowers['Mark'],label="Male's Lowest Marks",color='y')
plt.scatter(femaleLowers['Sl No'],femaleLowers['Mark'],label="Female's Lowest Marks",color='purple')
plt.legend(bbox_to_anchor = (1.5 , 1.03))
plt.show() 
print("Total mean mark :{:.2f}\n".format(mean_mark))
print("Male's mean mark :{:.2f}\nMale's highest mark :{}\nMale's lowest mark :{}\n"
      .format(maleMean,maleHigh,maleLow))
print("Female's mean mark :{:.2f}\nFemale's highest mark :{}\nFemale's lowest mark :{}\n"
      .format(femaleMean,femaleHigh,femaleLow))
if maleMean>femaleMean:
    print("Male's have better mean marks.")
else:
    print("Female's have better mean marks.")
```
