```python
import matplotlib.pyplot as plt
x=[1,4,2,5,8,6]#assign an array
plt.plot(x)#to plot the array
plt.title("Normal Graph")
plt.xlabel("X - axis")#label for x axis
plt.ylabel("Y - axis")#label for y axis
plt.grid()#to show gridlines
plt.show()#to hold the plot

plt.figure()
plt.title("Stem Graph")#to insert a new figure
plt.stem(x)#to visualize stem plot of the same array
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.show()
```
